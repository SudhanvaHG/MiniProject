Import the necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
mydata = pd.read_csv('imdb_data.csv',encoding = "utf-8")
```

```python
print(mydata.shape[0]) #number of rows
print(mydata.shape[1]) #number of columns
```

```python
# Select relevant features for clustering
X = mydata[['movie_year', 'rating']]
```

```python
# Convert genre column to numerical values using one-hot encoding
# genre_onehot = pd.get_dummies(X['director_name'])
# X = pd.concat([X, genre_onehot], axis=1)
# X.drop('director_name', axis=1, inplace=True)
```

```python
num_clusters = 3  # Number of clusters to form
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)
```

```python
mydata['cluster'] = kmeans.labels_
```

```python
cluster_centers = kmeans.cluster_centers_
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

for cluster_num in range(num_clusters):
    cluster_data = mydata[mydata['cluster'] == cluster_num]
    plt.scatter(cluster_data['rating'], cluster_data['movie_year'], color=colors[cluster_num], label=f'Cluster {cluster_num}')
    plt.scatter(cluster_centers[:, 1], cluster_centers[:, 0], color='black', marker='x', s=100, label='Cluster Centers')

plt.xlabel('Rating')
plt.ylabel('Genre')
plt.title('Movie Clusters')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data
```

```python
# Perform K-means clustering

n_clusters = 3
n_init = 10  # or 'auto'

kmeans = KMeans(n_clusters=n_clusters, n_init=n_init)
```

```python
kmeans.fit(data)
```

```python
cluster_centers = kmeans.cluster_centers_
```

```python
# Add cluster labels to the DataFrame
labels = kmeans.labels_
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
colors = ['r', 'g', 'b']
for cluster_num in range(num_clusters):
    cluster_data = data[labels == cluster_num]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], color=colors[cluster_num], label=f'Cluster {cluster_num}')
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='x', s=100, label='Cluster Centers')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Clusters with Cluster Centers')
plt.legend()
plt.show()
```

```python
print("Cluster Centers:")
for i, center in enumerate(cluster_centers):
    print(f"Cluster {i+1}: {center}")
```
